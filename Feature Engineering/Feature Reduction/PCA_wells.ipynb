```python
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.decomposition import PCA
```

### Read the data table

```python
df = pd.read_csv("unconv_MV.csv")
df.head()
```

```python
x_orig = df.loc[:, ['Por','LogPerm']].values
x = StandardScaler().fit_transform(x_orig)
```

```python
n_components = 2
pca = PCA(n_components=n_components)
pca.fit(x)

fig, ax = plt.subplots(1,3, figsize = (16,5))

ax[0].scatter(x[:,0], x[:,1], s=7, c="blue",  alpha=0.2, linewidths=1.0)
ax[0].set_xlabel('Standardized Por');
ax[0].set_ylabel('Standardized LogPerm')

x_trans = pca.transform(x)  
ax[1].scatter(x_trans[:,0], x_trans[:,1], s=7, c="red",  alpha=0.2, linewidths=1.0)
ax[1].set_xlabel('PC1');
ax[1].set_ylabel('PC2'); # We lose the physical meaning

x_back = pca.inverse_transform(x_trans)  
ax[2].scatter(x_back[:,0], x_back[:,1], s=7, c="blue",  alpha=0.2, linewidths=1.0)
ax[2].set_xlabel('Standardized Por');
ax[2].set_ylabel('Standardized LogPerm')

plt.show()
```

```python
n_components = 1
pca = PCA(n_components=n_components)
pca.fit(x)

fig, ax = plt.subplots(1,3, figsize = (16,5))

ax[0].scatter(x[:,0], x[:,1], s=7, c="blue",  alpha=0.2, linewidths=1.0)
ax[0].set_xlabel('Standardized Por');
ax[0].set_ylabel('Standardized LogPerm')

x_trans = pca.transform(x)  
ax[1].scatter(x_trans[:,0], np.zeros(len(x_trans)), s=7, c="red",  alpha=0.2, linewidths=1.0)
ax[1].set_xlabel('PC1');
ax[1].set_ylabel('PC2'); # We lose the physical meaning

x_back = pca.inverse_transform(x_trans)  
ax[2].scatter(x_back[:,0], x_back[:,1], s=7, c="blue",  alpha=0.2, linewidths=1.0)
ax[2].set_xlabel('Standardized Por');
ax[2].set_ylabel('Standardized LogPerm')

plt.show()
```
